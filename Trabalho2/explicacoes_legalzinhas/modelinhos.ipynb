# Regressao gamma

Da classe dos modelos generalizados, usa um preditor linear (igual regressao linear) para modelar uma funcao (funcao de ligacao) da media da resposta, e a varaivel resposta tem outra distribuição (a normal eh um caso espcial que se reduz a LinReg padrao). Foi utilizada uma resposta gamma com funcao de ligacao identidade.

$$
y_i \sim Gamma(\mu_i, \phi) \\
g(\mu_i) = \eta_i \\
\eta_i = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_p x_p \\
\text{Nesse caso:}\quad g(\mu_i) = \mu_i \quad\text{(Identidade)}\\
$$

Distribuicao gamma eh positiva, assimetrica a direita, com caudas pesadas e variancia proporcional a media, como o diagnostico do modelo normal mostrou sobredispersao e assimetria (residuos se espalham mais que a normal), a distribuicao gamma poderia ser adequada para resolver este problema

## Resultado

Teve menor MAR que o modelo linear, indica maior robustez,  mas maior MSE. Uso interativo (feira) mostrou possivel overfitting para aptos pequenos


Diagnostico revelou sobredspersao e assimetria e residuos que aparentavam ter maior variancia conforme a media, justificando o uso da regressao gamma, area foi variavel mto importante, closet, escritorio

# XGBoost

Usa arvores de decisao (weak learners) em conjunto para realizar predicoes, 

## Boosting

Boosting eh uma tecnica de ML onde weak learners (modelos fracos) sao treinados sequencialmente para obter melhores resultados

A cada passo (treino de modelo), a funcao de perda otimizada nao leva em conta apenas o modelo atual, mas tambem as predicoes dos modelos ja treinados, ou seja, cada weak learner tenta "corrigir" os erros dos modelos anteriores. As predicoes sao realizadas tomando a media (ponderada por um learning rate) dos modelos treinados ao fim

O XGBoost eh uma implementacao de gradient boosting, uma variante do boosting onde a expansao de taylor da funcao de erro eh utilizada, o que permite mais flexibilidade das funcoes de perda utilizadas.

O XGBoost eh altamente otimizado, possui uma das melhores precisoes para dados tabulares out of the box, suporta nativamente vairiaveis categoricas e valores faltantes, possui diversas opcoes de regularizacao, permite usar tanto arvores como modelos lineares como weak learners, etc, etc, etc o modelo so eh bom msm

Adicionamente, para ter modelos mais robustos, no boosting, cada arvore ve apenas um subconjunto dos dados, podendo ser configurado tanto em questao de linhas e colunas dos dados
